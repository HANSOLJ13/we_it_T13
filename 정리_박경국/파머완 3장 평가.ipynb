###정밀도
- TP / (FP + TP)
- positive 예측 성능 중요

###재현율
- TP / (FN + TP)
- positive를 negative로 판단하면 큰 영향 발생하는 경우

###precision / recall trade-off

###F1 score
- F1 = 2 × { (precision × recall) / (precision + recall) }
- precision과 recall 중 어느 한쪽으로 치우치지 않을 때 높음

###ROC AUC
- ROC곡선: FPR이 변할 때 TPR이 어떻게 변하는지 나타낸 곡선
- ROC 곡선 밑의 면적
