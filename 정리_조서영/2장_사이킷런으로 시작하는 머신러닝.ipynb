## ✨ 1. Scikit-Learn (사이킷런) 🧠

### 🏗️ Estimator 이해 및 주요 메서드
- **Estimator**: 머신러닝 모델을 의미 (ex: `LinearRegression`, `DecisionTreeClassifier` 등)
- **🛠️ fit()**: 모델 학습 (훈련 데이터 적용)
- **🔮 predict()**: 예측 수행 (새로운 데이터 입력)

### 🔑 사이킷런 주요 모듈
- 📊 `sklearn.datasets`: 내장된 예제 데이터 제공
- 🔄 `sklearn.model_selection`: 학습/테스트 분리, 교차 검증, 하이퍼파라미터 튜닝
- 🎨 `sklearn.preprocessing`: 데이터 전처리 (스케일링, 인코딩 등)
- 🏅 `sklearn.metrics`: 모델 평가 지표 제공
- 🌳 `sklearn.ensemble`: 앙상블 학습 기법 제공

### 📚 내장된 예제 데이터 세트
- `load_iris()`, `load_boston()`, `load_digits()` 등 다양한 샘플 데이터 제공

## 🚀 2. Model Selection (모델 선택 및 평가)

### 📏 학습/테스트 데이터 분리
- 데이터를 **학습(train) / 테스트(test)** 세트로 나누어 모델 성능 평가 🎯
- 예시: 시험 공부할 때, 일부 문제(학습 데이터)로 공부하고 나머지 문제(테스트 데이터)로 실력을 검증하는 것과 비슷해!

### 🔄 교차 검증 (Cross Validation)
- **🎳 K-Fold 교차 검증**: 데이터를 K개로 나눠 여러 번 학습/평가 반복
- **⚖️ Stratified K-Fold**: 불균형 데이터셋에서 클래스 비율 유지하면서 K-Fold 적용
- **📊 cross_val_score()**: 여러 번의 교차 검증 점수 평균 계산

✅ **쉽게 말하면?**
시험을 여러 번 다른 문제 조합으로 보고 평균 점수를 내는 것과 비슷해! 🎯

### 🛠️ GridSearchCV (하이퍼파라미터 튜닝)
- 교차 검증과 하이퍼파라미터 최적화를 동시에 수행 ✨
- **비유하자면?**
라면을 끓일 때 물의 양, 면의 삶는 시간, 스프의 양을 조절해가며 최적의 조합을 찾는 것과 같아! 🍜🔥

## 🧹 3. 데이터 전처리 (Data Preprocessing)

### 🔠 데이터 인코딩 (Encoding)
- **🏷️ Label Encoding**: 카테고리 데이터를 숫자로 변환
- **🎨 One-Hot Encoding**: 숫자로 변환한 데이터를 **이진 벡터**로 변환

🔎 **이진 벡터(Binary Vector)란?**
👉 0과 1로만 이루어진 벡터!

💡 **예시:**
```
카테고리 데이터: ['사과', '바나나', '포도', '바나나', '사과']

Label Encoding 결과: [0, 1, 2, 1, 0]  ❌ 숫자가 서열(순서) 정보를 포함하는 것처럼 보이는 문제

One-Hot Encoding 결과:
사과  → [1, 0, 0]  🍎
바나나 → [0, 1, 0]  🍌
포도  → [0, 0, 1]  🍇
```
**🎯 핵심: 숫자로 순위를 매기지 않고, 각 카테고리를 독립적으로 표현하는 방법!**

### 📏 피처 스케일링과 정규화 (Feature Scaling & Normalization)
- **📉 StandardScaler** (표준화): 평균 0, 분산 1로 변환
- **📊 MinMaxScaler** (정규화): 데이터 값을 0~1 범위로 조정

🧐 **언제 사용하면 좋을까?**
- **StandardScaler** → 키, 몸무게처럼 **평균을 중심으로 변동성이 중요한 데이터**
- **MinMaxScaler** → 성적(0~100점)처럼 **값의 범위가 중요한 데이터**

### ⚠️ 스케일링 시 유의점
- 학습 데이터(`fit`)와 테스트 데이터(`transform`)를 분리하여 적용해야 함 ⚠️
- **비유하자면?**
시험 문제(학습 데이터)만 보고 시험 범위(스케일)를 정해야지, 시험 본 후(테스트 데이터) 기준을 바꾸면 안 되는 것과 같아! 🎓📚
